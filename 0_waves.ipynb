# The Basics of Waves

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set()
%matplotlib inline
```

## Model a wave

```python
x = np.linspace(0, 20, 201)
y = np.sin(x)

plt.figure(figsize = (5, 3))
plt.plot(x, y, 'b')
plt.ylabel('Amplitude')
plt.xlabel('Location (x)')
plt.show()
```

```python
fig = plt.figure(figsize = (8,8))

times = np.arange(5)

n = len(times)

for t in times:
    plt.subplot(n, 1, t+1)
    y = np.sin(x + t)
    plt.plot(x, y, 'b')
    plt.plot(x[25], y [25], 'ro')
    plt.ylim(-1.1, 1.1)
    plt.ylabel('y')
    plt.title(f't = {t}')

plt.xlabel('location (x)')
plt.tight_layout()
plt.show()
```

## Characteristics of a wave

| Property | Description |
| --- | --- |
| Amplitude | The height of the wave |
| Frequency | The number of waves per second |
| Wavelength | The distance between two peaks |
|![period](image/24.01.01-period_amplitude.jpg) |![Wavelength](image/24.01.02-wavelength_amplitude.jpg)

$$
period = \frac{1}{frequency}
$$

$$
y(t) = Asin(\omega{t}+\phi)\quad \omega = \frac{2\pi}{T} = 2\pi{f}
$$

### change the frequency

```python
# sampling rate
sr = 100.0
# sampling interval
ts = 1.0/sr
t = np.arange(0,1,ts)

# frequency of the signal
freq = 5   
y = np.sin(2*np.pi*freq*t)

plt.figure(figsize = (8, 8))
plt.subplot(211)
plt.plot(t, y, 'b')
plt.ylabel('Amplitude')

freq = 10   
y = np.sin(2*np.pi*freq*t)

plt.subplot(212)
plt.plot(t, y, 'b')
plt.ylabel('Amplitude')

plt.xlabel('Time (s)')
plt.show()
```

### change the amplitude

```python
# frequency of the signal
freq = 5   
y = 5*np.sin(2*np.pi*freq*t)

plt.figure(figsize = (8, 8))
plt.subplot(211)
plt.plot(t, y, 'b')
plt.ylabel('Amplitude')

y = 10*np.sin(2*np.pi*freq*t + 10)

plt.subplot(212)
plt.plot(t, y, 'b')
plt.ylabel('Amplitude')

plt.xlabel('Time (s)')
plt.show()
```
